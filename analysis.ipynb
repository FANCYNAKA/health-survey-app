```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Administrator\Desktop\healthcare-survey-app\survey_data.csv")

df.head()
```

```python
# Convert income column to numbers
df["Income"] = pd.to_numeric(df["Income"])

# Sort data by income
highest_income = df.sort_values(by="Income", ascending=False)

# Create figure size
plt.figure(figsize=(10,5))

# Create bar chart
plt.bar(
    highest_income["Age"].astype(str),
    highest_income["Income"]
)

# Add chart title
plt.title("Ages with Highest Income")

# Label x-axis
plt.xlabel("Age")

# Label y-axis
plt.ylabel("Income")

# Display chart
plt.show()
```

```python
# Group by age and calculate average income
average_income = df.groupby("Age")["Income"].mean()

# Create chart size
plt.figure(figsize=(10,5))

# Create bar chart
average_income.plot(kind="bar")

# Add title and labels
plt.title("Average Income by Age")
plt.xlabel("Age")
plt.ylabel("Average Income")

# Show chart
plt.show()
```

```python
# Group spending by gender
gender_spending = df.groupby("Gender")[[
    "Utilities Amount",
    "Entertainment Amount",
    "School Fees Amount",
    "Shopping Amount",
    "Healthcare Amount"
]].sum()

# Create chart size
plt.figure(figsize=(12,6))

# Create grouped bar chart
gender_spending.plot(kind="bar")

# Add title and labels
plt.title("Gender Distribution Across Spending Categories")
plt.xlabel("Gender")
plt.ylabel("Amount Spent")

# Show chart
plt.show()
```

```python
# Create average income chart again
average_income.plot(kind="bar", figsize=(10,5))

# Add labels
plt.title("Average Income by Age")
plt.xlabel("Age")
plt.ylabel("Average Income")

# Save chart as image
plt.savefig("income_chart.png")

# Show chart
plt.show()
```

```python
# Create gender spending chart again
gender_spending.plot(kind="bar", figsize=(12,6))

# Add labels
plt.title("Gender Distribution Across Spending Categories")
plt.xlabel("Gender")
plt.ylabel("Amount Spent")

# Save chart
plt.savefig("gender_spending_chart.png")

# Show chart
plt.show()
```
